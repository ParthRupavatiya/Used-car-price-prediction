# Car Data from OLX

```python
# Scrape car data for mumbai location

driver = webdriver.Chrome("I:\Software//chromedriver.exe")

# open url in driver
driver.get("https://www.olx.in/")

driver.maximize_window() # maximize driver window

driver.implicitly_wait(10) # wait for open all elements of the page
```

```python
# click on buy car
driver.find_element_by_xpath("//*[@id='container']/main/div/section/div[1]/div/div/button[1]").click()
time.sleep(8)  # sleep code for 8 second

# clear search field
driver.find_element_by_xpath("//*[@id='container']/header/div/div/div[2]/div/div/div[1]/div/div[1]/input").clear()

# type mumbai in search field
driver.find_element_by_xpath("//*[@id='container']/header/div/div/div[2]/div/div/div[1]/div/div[1]/input").send_keys("mumbai")

# click on mumbai,maharashtra
driver.find_element_by_xpath("//*[@id='container']/header/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div[1]/div/span").click()

# Click on maruti suzuki brand filter
driver.find_element_by_xpath("//*[@id='c-make-maruti-suzuki']").click()
time.sleep(8)

# click on petrol car filter
driver.find_element_by_xpath("//*[@id='f-petrol-petrol']").click()
```

```python
# load more pages by clicking on load more button
count=15
while count>1:
    button=driver.find_element_by_xpath("//li[@class='_20FqS']/div/button")
    button.click()
    count-=1
    time.sleep(2)
```

```python
location=[]
car_model=[]
manuf_year=[]
driven_kilometers=[]
price=[]

# collecting locations
location_ = driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[3]/ul/li/a/div/div/span[1]") 
for i in location_:                        
    if i.text is None:
        location.append("--") # appending the "--" if data is not is there
    else:
        location.append(i.text)

# collecting car model
car_model_ = driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[3]/ul/li/a/div/span[3]") 
for i in car_model_:
    if i.text is None:
        car_model.append("--") # appending the "--" if data is not is there
    else:
        car_model.append(i.text)
    
    
# collecting manufacturing years   
manuf_year_ = driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[3]/ul/li/a/div/span[2]") 
for i in manuf_year_:
    if i.text is None:
        manuf_year.append("--") # appending the "--" if data is not is there
    else:
        manuf_year.append(i.text)
    
    
# collecting driven kilometers    
driven_kilometers_ = driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[3]/ul/li/a/div/span[2]") 
for i in driven_kilometers_:
    if i.text is None:
        driven_kilometers.append("--") # appending the "--" if data is not is there
    else:
        driven_kilometers.append(i.text)
        
    
# collecting car price  
price_ = driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[3]/ul/li/a/div/span[1]") 
for i in price_:                      
    if i.text is None:
        price.append("--") # appending the "--" if data is not is there
    else:
        price.append(i.text)
        
print(location)
print(car_model)
print(manuf_year)
print(driven_kilometers)
print(price)
```

```python
# remove extra information from the location
locations_=[]
for x in location:
    locations_.append(x.split(",")[1])
print(locations_)

# remove extra information from the manufacturing year
mnf_year=[]
for x in manuf_year:
    mnf_year.append(x.split(" -")[0])
print(mnf_year)

# remove extra information from the driven kilometers
driven_km=[]
driven_kms=[]
for x in driven_kilometers:
    driven_km.append(x.split("km")[0])
for x in driven_km:
    driven_kms.append(x.split("- ")[1])
print(driven_kms)

# remove extra information from the price of car
Price=[]
for x in price:
    Price.append(x.split("â‚¹ ")[1])
print(Price)
```

```python
# close driver
driver.close()
```

```python
# during scraping data, we fiter brand and fuel.
# So, instead of extracting those features we manually add those columns in our data.
# make dataframe
df1 = pd.DataFrame({})
df1["Location"] = locations_
df1["Brand"] = "MARUTI"
df1["Model"] = car_model
df1["Mnf_year"] = mnf_year
df1["Driven_kms"] = driven_kms
df1["Fuel"] = "Petrol"
df1["Car_Price"] = Price
df1
```

```python
# Scrape car data for all india location

driver = webdriver.Chrome("I:\Software//chromedriver.exe")

# open url in driver
driver.get("https://droom.in/cars/used?page=1&tab=grid&display_category=All+Cars&condition=used")

driver.maximize_window() # maximize driver window

driver.implicitly_wait(10) # wait for open all elements of the page
```

```python
# load more pages by clicking on load more button
count=90
while count>1:
    button=driver.find_element_by_xpath("//*[@id='loadMore']/button")
    button.click()
    count-=1
    time.sleep(2) # sleep code for 2 seconds
```

```python
# create empty lists for store scraped data
location=[]
brand=[]
car_model=[]
manuf_year=[]
driven_kilometers=[]
fuel=[]
price=[]

# collecting locations
location_ = driver.find_elements_by_xpath("//*[@id='search_results']/div/section/div/div/div/div[4]/div[4]/div/div[4]/label") 
for i in location_:                        
    if i.text is None:
        location.append("--") # appending the "--" if data is not is there
    else:
        location.append(i.text)
        
# collecting brands
brand_ = driver.find_elements_by_xpath("//*[@id='search_results']/div/section/div/div/div/div[4]/div[1]/h4") 
for i in brand_:                    
    if i.text is None:
        brand.append("--") # appending the "--" if data is not is there
    else:
        brand.append(i.text)

# collecting car model
car_model_ = driver.find_elements_by_xpath("//*[@id='search_results']/div/section/div/div/div/div[4]/div[1]/h4") 
for i in car_model_:
    if i.text is None:
        car_model.append("--") # appending the "--" if data is not is there
    else:
        car_model.append(i.text)
    
    
# collecting manufacturing years   
manuf_year_ = driver.find_elements_by_xpath("//*[@id='search_results']/div/section/div/div/div/div[4]/div[1]/h4") 
for i in manuf_year_:
    if i.text is None:
        manuf_year.append("--") # appending the "--" if data is not is there
    else:
        manuf_year.append(i.text)
    
    
# collecting driven kilometers    
driven_kilometers_ = driver.find_elements_by_xpath("//*[@id='search_results']/div/section/div/div/div/div[4]/div[4]/div/div[1]/label") 
for i in driven_kilometers_:                        
    if i.text is None:
        driven_kilometers.append("--") # appending the "--" if data is not is there
    else:
        driven_kilometers.append(i.text)
        
        
# collecting fuel type  
fuel_ = driver.find_elements_by_xpath("//*[@id='search_results']/div/section/div/div/div/div[4]/div[4]/div/div[3]/label") 
for i in fuel_:                        
    if i.text is None:
        fuel.append("--") # appending the "--" if data is not is there
    else:
        fuel.append(i.text)
        
    
# collecting car price  
price_ = driver.find_elements_by_xpath("//*[@id='search_results']/div/section/div/div/div/div[4]/div[2]")
for i in price_:                        
    if i.text is None:
        price.append("--") # appending the "--" if data is not is there
    else:
        price.append(i.text)
        
print(location)
print(brand)
print(car_model)
print(manuf_year)
print(driven_kilometers)
print(fuel)
print(price)
```

```python
# remove extra information from the brand 
brands=[]
for x in brand:
    brands.append(x.split(" ")[0])
print(brands)

# remove extra information from the car model
car_models=[]
for x in car_model:
    car_models.append(x[0:-5])
print(car_models)


# remove extra information from the manufacturing year
mnf_year=[]
for x in manuf_year:
    mnf_year.append(x.split(" ")[-1])
print(mnf_year)

# remove old price and take new price
prices=[]
for x in price:
    prices.append(x.split(" ")[-1])
print(prices)

# remove extra information from the driven kilometers
driven_kms=[]
for x in driven_kilometers:
    driven_kms.append(x.split(" Km")[0])
print(driven_kms)
```
